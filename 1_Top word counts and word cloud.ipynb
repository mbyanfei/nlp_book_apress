# Get frequency distribution of words on removing stop words

```python
def remove_stop_words(words,st):
    new_list = []
    st = set(st)
    for i in words:
        if(i not in st):
            new_list.append(i)
            
    return new_list
```

```python
list_all1 = remove_stop_words(words,st)
##we did not word_tokenize as the returned object is already in a list form
fdist = FreqDist(list_all1)
```

```python
fdist.most_common()
```

# Get Wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def generate_wordcloud(text): 
    wordcloud = WordCloud(
            background_color ='white',relative_scaling = 1,
                          collocations = False,
                          ).generate(text)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    
    return wordcloud
```

```python
##joining the list to a string as an input to wordcloud
str_all_rejoin = ' '.join(list_all1)
```
